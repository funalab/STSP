```python
import os
import random
from glob import glob
import numpy as np

np.random.seed(42)
random.seed(42)
```

```python
path_img = glob('/Users/tokkuman/git-tokkuman/seminiferous_tubule_stage_classification/datasets/images/*.tif')
with open('/Users/tokkuman/git-tokkuman/seminiferous_tubule_stage_classification/datasets/label_list.txt', 'r') as f:
    label_list = [line.rstrip() for line in f]

label_list_each_mouse = {
    1: ['20210709'],
    2: ['20211015-RR-2'],
    3: ['20211015-RR-3'],
    4: ['20211015 R'],
    5: ['211022'],
    6: ['211027'],
    7: ['211027_z'],
    8: ['20211029'],
    9: ['20211108_a'],
    10: ['20211109_a'],
    11: ['20211109_b', '20211110_d', '20211115_d'],
    12: ['20211109_c', '20211110_e', '20211111_a', '20211111_b'],
    13: ['20211110_a', '20211111_c'],
    14: ['20211110_b', '20211111_d', '20211112_a', '20211115_a', '20211115_b'],
    15: ['20211110_c', '20211115_c'],
    16: ['B6N_#1'],
    17: ['B6N_#2'],
    18: ['B6N_#3'],
    19: ['BDF1_#1'],
    20: ['BDF1_#2'],
    21: ['BDF1_#3'],
}

path_cls = {}
for i in label_list:
    path_cls[i] = []
for p in path_img:
    cls = os.path.basename(p)[os.path.basename(p).rfind('_')+1:os.path.basename(p).rfind('-')]
    if cls in label_list:
        path_cls[cls].append(p)

total = 0
for i in label_list:
    print(i, len(path_cls[i]))
    random.shuffle(path_cls[i])
    total += len(path_cls[i])
print('total: {}'.format(total))
```

```python
group_list = [i for i in range(1, 22)]
random.shuffle(group_list)
fold = 4
root = '/Users/tokkuman/git-tokkuman/seminiferous_tubule_stage_classification/datasets'
split_path = 'split_list_each_mouse_datasets_20220922'
os.makedirs(os.path.join(root, split_path), exist_ok=True)

os.makedirs(os.path.join(root, split_path, 'fold1'), exist_ok=True)
os.makedirs(os.path.join(root, split_path, 'fold2'), exist_ok=True)
os.makedirs(os.path.join(root, split_path, 'fold3'), exist_ok=True)
os.makedirs(os.path.join(root, split_path, 'fold4'), exist_ok=True)
print(group_list)
print(len(path_img))

test_group = []
for i in range(0, 5):
    test_group += label_list_each_mouse[group_list[i]]
fold1_group, fold2_group, fold3_group, fold4_group = [], [], [], []
for i in range(5, 9):
    fold1_group += label_list_each_mouse[group_list[i]]
for i in range(9, 13):
    fold2_group += label_list_each_mouse[group_list[i]]
for i in range(13, 17):
    fold3_group += label_list_each_mouse[group_list[i]]
for i in range(17, 21):
    fold4_group += label_list_each_mouse[group_list[i]]

cnt = 0
fold1, fold2, fold3, fold4 = 0, 0, 0, 0
for p in path_img:
    lab = os.path.basename(p)[:os.path.basename(p).rfind('_')]
    if lab[:3] == "BDF" or lab[:3] == "B6N":
        lab = lab[:lab.rfind('_')]
    if lab in test_group:
        cnt += 1
        with open(os.path.join(root, split_path, 'test.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))

    elif lab in fold1_group:
        fold1 += 1
        with open(os.path.join(root, split_path, 'fold1', 'validation.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, split_path, 'fold2', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, split_path, 'fold3', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, split_path, 'fold4', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
            
    elif lab in fold2_group:
        fold2 += 1
        with open(os.path.join(root, split_path, 'fold1', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, split_path, 'fold2', 'validation.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, split_path, 'fold3', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, split_path, 'fold4', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
            
    elif lab in fold3_group:
        fold3 += 1
        with open(os.path.join(root, split_path, 'fold1', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, split_path, 'fold2', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, split_path, 'fold3', 'validation.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, split_path, 'fold4', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))

    elif lab in fold4_group:
        fold4 += 1
        with open(os.path.join(root, split_path, 'fold1', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, split_path, 'fold2', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, split_path, 'fold3', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, split_path, 'fold4', 'validation.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))

    else:
        print(lab)

print(cnt)
print(fold1, fold2, fold3, fold4)
print(cnt + fold1 + fold2 + fold3 + fold4)
```

```python
# trainとvalidationの被りなしを確認
for fd in range(0, fold):
    target = os.path.join(root, 'split_list_each_mouse', 'fold{}'.format(fd+1))
    with open(os.path.join(target, 'train.txt'), 'r') as f:
        filelist_train = [line.rstrip() for line in f]
    with open(os.path.join(target, 'validation.txt'), 'r') as f:
        filelist_validatioin = [line.rstrip() for line in f]
    print(len(np.unique(filelist_train + filelist_validatioin)))
    
filelist_validatioin_total = []
for fd in range(0, fold):
    target = os.path.join(root, 'split_list_each_mouse', 'fold{}'.format(fd+1))
    with open(os.path.join(target, 'validation.txt'), 'r') as f:
        filelist_validatioin = [line.rstrip() for line in f]
    filelist_validatioin_total += filelist_validatioin
print(len(np.unique(filelist_validatioin_total)))
```
