```python
import os
import random
from glob import glob
import numpy as np

np.random.seed(0)
random.seed(0)
```

```python
path_img = glob('/Users/tokkuman/git-tokkuman/seminiferous_tubule_stage_classification/datasets/images/*.tif')
with open('/Users/tokkuman/git-tokkuman/seminiferous_tubule_stage_classification/datasets/label_list.txt', 'r') as f:
    label_list = [line.rstrip() for line in f]

label_list_each_mouse = {
    1: ['20210709'],
    2: ['20211015-RR-2'],
    3: ['20211015-RR-3'],
    4: ['20211015 R'],
    5: ['211022'],
    6: ['211027'],
    7: ['211027_z'],
    8: ['20211029'],
    9: ['20211108_a'],
    10: ['20211109_a'],
    11: ['20211109_b', '20211110_d', '20211115_d'],
    12: ['20211109_c', '20211110_e', '20211111_a', '20211111_b'],
    13: ['20211110_a', '20211111_c'],
    14: ['20211110_b', '20211111_d', '20211112_a', '20211115_a', '20211115_b'],
    15: ['20211110_c', '20211115_c']
}

path_cls = {}
for i in label_list:
    path_cls[i] = []
for p in path_img:
    cls = os.path.basename(p)[os.path.basename(p).rfind('_')+1:os.path.basename(p).rfind('-')]
    if cls in label_list:
        path_cls[cls].append(p)

total = 0
for i in label_list:
    print(i, len(path_cls[i]))
    random.shuffle(path_cls[i])
    total += len(path_cls[i])
print('total: {}'.format(total))
```

```python
group_list = [i for i in range(1, 16)]
random.shuffle(group_list)
fold = 4
root = '/Users/tokkuman/git-tokkuman/seminiferous_tubule_stage_classification/datasets'
os.makedirs(os.path.join(root, 'split_list_each_mouse'), exist_ok=True)
os.makedirs(os.path.join(root, 'split_list_each_mouse/fold1'), exist_ok=True)
os.makedirs(os.path.join(root, 'split_list_each_mouse/fold2'), exist_ok=True)
os.makedirs(os.path.join(root, 'split_list_each_mouse/fold3'), exist_ok=True)
os.makedirs(os.path.join(root, 'split_list_each_mouse/fold4'), exist_ok=True)
print(group_list)
print(len(path_img))

cnt = 0
fold1, fold2, fold3, fold4 = 0, 0, 0, 0
for p in path_img:
    lab = os.path.basename(p)[:os.path.basename(p).rfind('_')]
    if lab in label_list_each_mouse[group_list[0]] + label_list_each_mouse[group_list[1]] + label_list_each_mouse[group_list[2]]:
        cnt += 1
        with open(os.path.join(root, 'split_list_each_mouse', 'test.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))

    elif lab in label_list_each_mouse[group_list[3]] + label_list_each_mouse[group_list[4]] + label_list_each_mouse[group_list[5]]:
        fold1 += 1
        with open(os.path.join(root, 'split_list_each_mouse', 'fold1', 'validation.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, 'split_list_each_mouse', 'fold2', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, 'split_list_each_mouse', 'fold3', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, 'split_list_each_mouse', 'fold4', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
            
    elif lab in label_list_each_mouse[group_list[6]] + label_list_each_mouse[group_list[7]] + label_list_each_mouse[group_list[8]]:
        fold2 += 1
        with open(os.path.join(root, 'split_list_each_mouse', 'fold1', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, 'split_list_each_mouse', 'fold2', 'validation.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, 'split_list_each_mouse', 'fold3', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, 'split_list_each_mouse', 'fold4', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
            
    elif lab in label_list_each_mouse[group_list[9]] + label_list_each_mouse[group_list[10]] + label_list_each_mouse[group_list[11]]:
        fold3 += 1
        with open(os.path.join(root, 'split_list_each_mouse', 'fold1', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, 'split_list_each_mouse', 'fold2', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, 'split_list_each_mouse', 'fold3', 'validation.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, 'split_list_each_mouse', 'fold4', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))

    elif lab in label_list_each_mouse[group_list[12]] + label_list_each_mouse[group_list[13]] + label_list_each_mouse[group_list[14]]:
        fold4 += 1
        with open(os.path.join(root, 'split_list_each_mouse', 'fold1', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, 'split_list_each_mouse', 'fold2', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, 'split_list_each_mouse', 'fold3', 'train.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))
        with open(os.path.join(root, 'split_list_each_mouse', 'fold4', 'validation.txt'), 'a') as f:
            f.write('{}\n'.format(os.path.basename(p)))

    else:
        print(lab)

print(cnt)
print(fold1, fold2, fold3, fold4)
print(cnt + fold1 + fold2 + fold3 + fold4)
# with open(os.path.join(root, 'split_list_each_mouse', 'test.txt'), 'w') as f:
#     for p in path_img:
#         for i in group_list[:3]:
#             if 
#             f.write('{}\n'.format(os.path.basename(path_cls[c][0])))
#             path_cls[c].pop(0)
       
# total = 0
# for i in label_list:
#     total += len(path_cls[i])
# print('total (train+validation): {}'.format(total))        

# for fd in range(0, fold):
#     dist = os.path.join(root, 'split_list_each_mouse', 'fold{}'.format(fd+1))
#     os.makedirs(dist, exist_ok=True)
#     with open(os.path.join(dist, 'train.txt'), 'w') as f:
#         for c in label_list:
#             #print(len(path_cls[c][:int(len(path_cls[c])/fold)*fd] + path_cls[c][int(len(path_cls[c])/fold)*(fd+1):]))
#             for pa in path_cls[c][:int(len(path_cls[c])/fold)*fd] + path_cls[c][int(len(path_cls[c])/fold)*(fd+1):]:
#                 f.write('{}\n'.format(os.path.basename(pa)))

#     with open(os.path.join(dist, 'validation.txt'), 'w') as f:
#         for c in label_list:
#             for pa in path_cls[c][int(len(path_cls[c])/fold)*fd:int(len(path_cls[c])/fold)*(fd+1)]:
#                 f.write('{}\n'.format(os.path.basename(pa)))
```

```python
# trainとvalidationの被りなしを確認
for fd in range(0, fold):
    target = os.path.join(root, 'split_list_each_mouse', 'fold{}'.format(fd+1))
    with open(os.path.join(target, 'train.txt'), 'r') as f:
        filelist_train = [line.rstrip() for line in f]
    with open(os.path.join(target, 'validation.txt'), 'r') as f:
        filelist_validatioin = [line.rstrip() for line in f]
    print(len(np.unique(filelist_train + filelist_validatioin)))
    
filelist_validatioin_total = []
for fd in range(0, fold):
    target = os.path.join(root, 'split_list_each_mouse', 'fold{}'.format(fd+1))
    with open(os.path.join(target, 'validation.txt'), 'r') as f:
        filelist_validatioin = [line.rstrip() for line in f]
    filelist_validatioin_total += filelist_validatioin
print(len(np.unique(filelist_validatioin_total)))
```
